## Authenticating

```python
BUCKET_NAME = "codefile-examples-bucket"  # @param {type:"string"}
PST_FILE = "Outlook.pst" # @param {type:"string"}
```

```python
import io
import os
from google.cloud import storage
```

## Leer archivo desde Cloud Storage

Acceder al Cloud Storage y descargar archivo .pst

```python
from pydrive.files import MIME_TYPE_TO_BOM
# Authenticate to Google Cloud
storage_client = storage.Client()

# Get the bucket object
bucket = storage_client.bucket(BUCKET_NAME)

# Get the blob object
blob = bucket.blob(PST_FILE)

# Download the blob to a local file
blob.download_to_filename(PST_FILE)

# Get the list of files in the /content folder
files = os.listdir("/content")
```

```python
path= "/content/"+PST_FILE
```

Analizar archivo pst

```python
import aspose.email
import aspose.email.storage.pst

# Load PST file
personalStorage = aspose.email.storage.pst.PersonalStorage.from_file(path)

# Get folders' collection
folderInfoCollection = personalStorage.root_folder.get_sub_folders()
```

```python
folderInfoCollection[0].display_name

```

```python
# Extract folders' information
for folderInfo in folderInfoCollection:
	print("Folder: " + folderInfo.display_name)
	print("Total Items: " + str(folderInfo.content_count))
	print("Total Unread Items: " + str(folderInfo.content_unread_count))
```

Extraer mensajes de un archivo pst

```python
# Load PST file
sourcePst = aspose.email.storage.pst.PersonalStorage.from_file(path)

# Select desired folder
sourceFolder = sourcePst.root_folder.get_sub_folder("Inbox")

# Get content in the folder
messageInfoCollection = sourceFolder.get_contents()


for messageInfo in messageInfoCollection:
    mapi = personalStorage.extract_message(messageInfo)
    print("Subject: " + mapi.subject)
    print("Sender name: " + mapi.sender_name)
    print("Sender email address: " + mapi.sender_email_address)
    print("To: ", mapi.display_to)
    print("Cc: ", mapi.display_cc)
    print("Bcc: ", mapi.display_bcc)
    print("Delivery time: ", str(mapi.delivery_time))
    print("Body: " + mapi.body)
```
